```python
splits=[]
```

```python
for ind in df.index: 
    if(pd.isna(df['Name'][ind])):
        splits.append(ind)
```

```python
splits
```

```python
dfs=[]
```

```python
for i in range(len(splits) -1):
    current_df = df[splits[i]:splits[i+1]]
    current_df.reset_index(drop=True, inplace=True)
    dfs.append(current_df)
```

```python
for ind in dfs[0].index:
    if(ind == 0):
        print(getCompanyName(df["Code"][ind]))
```

```python
def getCompanyName(name:str):
    return name.split(':')[-1]
```

```python
def getCompanyFromDf(adf:pd.core.frame.DataFrame):
    for ind in adf.index:
        if(ind == 0):
            return getCompanyName(adf["Code"][ind])
```

```python
for adf in dfs:
    print(getCompanyFromDf(adf))
    adf = adf[1:]
```
